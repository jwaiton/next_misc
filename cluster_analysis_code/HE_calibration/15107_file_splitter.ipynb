- Collect the DSTs and RECO for each of those groups

```python
def split_events(df, N):
    '''
    Splits dataframe by number of events into chunks each containing
    approximately N events per grouping.
    Parameters
    ----------
    df : pandas.DataFrame
        The dataframe containing the events from DST/Events
    N : int 
        The number of events per group.
    Returns
    -------
    list
        A list of lists, each containing ~N events.
    '''

    events = df.event.unique()
    num_events = len(events)
    num_groups = num_events // N
    print(f'{num_events} events split into {num_groups} groups')
    return np.array_split(events, num_groups)
```

```python
# select these events from DST/Events, RECO/Events and Run/events
# also make sure Run/runInfo is the same length of DST/Events
split_dir = 'files/15107_split/'
compression = 'ZLIB4'
for i, events in tqdm(enumerate(split_arr)):
    # select DF from dst with matching 'event' category
    chunked_DST  = dst_Events[dst_Events['event'].isin(events)]
    chunked_RECO    = RECO_Events[RECO_Events['event'].isin(events)]
    chunked_Runevts = Run_events[Run_events['evt_number'].isin(events)]
    chunked_RunInfo = Run_runInfo.head(len(events))
    

    with tb.open_file(split_dir + str('beer_') + str(i) + str('.h5'), 'w', filters = tbl.filters(compression)) as h5out:
        df_writer(h5out,   chunked_DST, "DST", "Events" , compression="ZLIB4")
        df_writer(h5out,   chunked_RECO, "RECO", "Events" , compression="ZLIB4")
        df_writer(h5out,   chunked_RunInfo, "Run" , "runInfo" , compression="ZLIB4")
        df_writer(h5out,   chunked_Runevts, "Run" , "events", compression="ZLIB4")
    #break
```
