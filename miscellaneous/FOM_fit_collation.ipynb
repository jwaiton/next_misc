```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
import pandas as pd
# import stuff
import sys,os,os.path

sys.path.append("/home/e78368jw/Documents/NEXT_CODE/next_misc/")
import core.functions as func
```

```python
# read in MC data FOM
fom_MC_1a = np.load('../FOM_merge&fit/data/fom_MC_isaura_1a.npy')
fom_MC_1a_err = np.load('../FOM_merge&fit/data/fom_MC_err_isaura_1a.npy')
print(len(fom_MC_1a))
cut_list = np.linspace(0,0.35, 36)

fom_MC_1b = np.load('../FOM_merge&fit/data/fom_MC_isaura_1b.npy')
fom_MC_1b_err = np.load('../FOM_merge&fit/data/fom_MC_err_isaura_1b.npy')

fom_MC_2a = np.load('../FOM_merge&fit/data/fom_MC_isaura_2a.npy')
fom_MC_2a_err = np.load('../FOM_merge&fit/data/fom_MC_err_isaura_2a.npy')

fom_MC_2b = np.load('../FOM_merge&fit/data/fom_MC_isaura_2b.npy')
fom_MC_2b_err = np.load('../FOM_merge&fit/data/fom_MC_err_isaura_2b.npy')

fom_MC_merge = np.load('../FOM_merge&fit/data/fom_MC_isaura_merged.npy')
fom_MC_merge_err = np.load('../FOM_merge&fit/data/fom_MC_err_isaura_merged.npy')
```

```python
#port_label = ['1A', '1B', '2A', '2B']
port_label = ['1A', '1B', '2A', '2B', 'ALL']
#fom_list = [fom_MC_1a, fom_MC_1b, fom_MC_2a, fom_MC_2b]
fom_list = [fom_MC_1a, fom_MC_1b, fom_MC_2a, fom_MC_2b, fom_MC_merge]
#fom_err = [fom_MC_1a_err, fom_MC_1b_err, fom_MC_2a_err, fom_MC_2b_err]
fom_err = [fom_MC_1a_err, fom_MC_1b_err, fom_MC_2a_err, fom_MC_2b_err, fom_MC_merge_err]
#linestyle = ['solid', 'dashed', 'dotted', 'dashdot']
linestyle = ['solid', 'dashed', 'dotted', 'dashdot', 'solid']
#colours = ['blue', 'y', 'red', 'green']
colours = ['blue', 'y', 'red', 'green', 'black']

plt.rcParams["figure.figsize"] = (10,10)
for i in range(len(fom_list)):
    # collect port name (using convention)
    # plot fom
    if (i == 4):
        opac = 1
    else:
        opac = 0.2
    plt.errorbar(cut_list, fom_list[i], fmt = 'o-', yerr=fom_err[i], label = "Port " + port_label[i], color = colours[i], alpha = opac, linestyle = linestyle[i])
    plt.title("NEXT100 MC - LPR - 208Tl Calibration", fontsize = 15)
    plt.xlabel("Blob-2 energy threshold (MeV)", fontsize = 13)
    plt.legend(loc='upper left', fontsize = 13, handlelength = 3)
    plt.xticks(fontsize = '12')
    plt.yticks(fontsize = '12')
    plt.ylabel("fom", fontsize = 13)
    print(str(port_label[i]) + " FOM maximum: " + str(max(fom_list[i])))

plt.grid()
plt.show()
```
