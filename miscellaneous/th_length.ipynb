```python
x = load_dsts(q, 'Tracking', 'Tracks')
```

```python
display(x)
```

```python
x.columns
```

```python
from tqdm import tqdm
v_d = 1.05 # mm per microsecond
t_length = []
drift_time  = []
# strip position and length from most energetic tracks and plot
for evt, df in tqdm(x.groupby('event')):
    #print(evt)
    #display(df)
    drift_time.append(v_d/df.iloc[0].z_ave)
    t_length.append(df.iloc[0].length)
    #print(drift_time)
    #print(t_length)

    
```

```python
import matplotlib.pyplot as plt
plt.scatter(drift_time, t_length)
plt.xlabel('Drift time ($\mu s$)')
plt.ylabel("Track length")
```

```python
qqq = np.array([drift_time, t_length])
print(qqq)
np.save('test', qqq)
```
