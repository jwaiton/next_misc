```python
file_name = '/home/e78368jw/Documents/NEXT_CODE/NEXUS/nexus/ICS_test.h5'

part_df = pd.read_hdf(file_name, 'MC/particles')
primaries = part_df[part_df.primary == True]
print(primaries.initial_volume.value_counts())
print(len(part_df[part_df.primary == True]))
```

```python
x = primaries.initial_x
y = primaries.initial_y
z = primaries.initial_z

print('X min = {0}, max = {1}'.format(min(x), max(x)))
print('Y min = {0}, max = {1}'.format(min(y), max(y)))
print('Z min = {0}, max = {1}'.format(min(z), max(z)))
```

```python
r_max = max(np.sqrt(x*x + y*y))
print(f'R max = {r_max}')
```

```python
offset = 20
xmin = min(x) - offset; xmax = max(x) + offset; ymin = min(y) - offset; ymax = max(y) + offset; 
zmin = min(z) - offset; zmax = max(z) + offset
bins = 100



h1 = plt.hist2d(x, y, bins=(bins, bins), range=((xmin, xmax), (ymin, ymax)), cmin=1)
#h1 = plt.hist2d(x, y, bins=(100, 100), range=((xmin, xmin +10), (ymin, ymax)), cmin=1)
#h1 = plt.hist2d(x, y, bins=(100, 100), range=((xmax-10, xmax), (ymin, ymax)), cmin=1)
plt.xlabel('x (mm)');
plt.ylabel('y (mm)');
#plt.ylim(280, 330)
#plt.xlim(-730, -600)
plt.colorbar()
plt.grid(True)
plt.show()

h2 = plt.hist2d(x, z, bins=(bins, bins), range=((xmin, xmax), (zmin, zmax)), cmin=1)
plt.xlabel('x (mm)');
plt.ylabel('z (mm)');
#plt.xlim(-350, -200)
plt.colorbar()
plt.grid(True)
plt.show()

h3 = plt.hist2d(z, y, bins=(bins, bins), range=((zmin, zmax), (ymin, ymax)), cmin=1)
plt.xlabel('z (mm)');
plt.ylabel('y (mm)');
#plt.ylim(260, 330)
#plt.xlim(655, 680)
plt.colorbar()
plt.grid(True)
plt.show()
```
