# Quick notebook to check the validity of scaling the space to be spherical in nature


- Create points within an oblate sphere

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters for the oblate sphere
a = 5  # Equatorial radius (XY in our case)
c = 2  # Polar radius (Z in our case)

offset_xy = 0
offset_z  = 0

# Generate random points
num_points = 1000
phi = np.random.uniform(0, 2 * np.pi, num_points)  # Azimuthal angle
theta = np.arccos(np.random.uniform(-1, 1, num_points))  # Polar angle

# Parametric equations for the oblate sphere
x = offset_xy+(a * np.sin(theta) * np.cos(phi))
y = offset_xy+(a * np.sin(theta) * np.sin(phi))
z = offset_z+(c * np.cos(theta))

```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.patches import Circle


def plot_xyz(x,y,z, a, c, offset_xy, offset_z):

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Scatter plot of the points
    ax.scatter(x, y, z, alpha=0.6)

    # Label axes
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    plt.show()

    # XZ projection
    fig, ax = plt.subplots(1, 3, figsize=(24, 8))

    circle0 = Circle((offset_xy, offset_xy), a, color='r', fill=False)
    circle1 = Circle((offset_xy, offset_z), a, color='r', fill=False)
    circle2 = Circle((offset_xy, offset_z), a, color='r', fill=False)
    #plt.show()

    # XY projection

    ax[0].add_patch(circle0)
    ax[0].scatter(x, y, alpha=0.6)
    ax[0].set_xlabel('X')
    ax[0].set_ylabel('Y')
    ax[0].set_title('XY Projection')
    ax[0].set_ylim([-a + offset_xy, a + offset_xy])
    ax[0].set_xlim([-a + offset_xy, a + offset_xy])

    ax[1].add_patch(circle1)
    ax[1].scatter(x, z, alpha=0.6)
    ax[1].set_xlabel('X')
    ax[1].set_ylabel('Z')
    ax[1].set_title('XZ Projection')
    ax[1].set_ylim([-a + offset_xy, a + offset_z])
    ax[1].set_xlim([-a + offset_xy, a + offset_z])


    # YZ projection

    ax[2].add_patch(circle2)
    ax[2].scatter(y, z, alpha=0.6)
    ax[2].set_xlabel('Y')
    ax[2].set_ylabel('Z')
    ax[2].set_title('YZ Projection')
    ax[2].set_ylim([-a + offset_xy, a + offset_z])
    ax[2].set_xlim([-a + offset_xy, a + offset_z])

    plt.show()
```

```python
plot_xyz(x,y,z, a, c, offset_xy, offset_z)
```

# projection scaling

alter Z by factor 2

```python
b = a/c # scale factor

Z = z*b

plot_xyz(x,y,Z, a, c, offset_xy, offset_z)
```
