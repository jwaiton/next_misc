```python
from invisible_cities.core.core_functions import in_range
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10000

x_array = np.random.rand(N)
print(x_array)
```

```python
y_array = np.random.normal(x_array * 10 + 5, scale = 1)
```

```python
down = lambda x: x * 11 + 3
up = lambda x: x *9 + 6

x_space = np.linspace(0, 1, 100)
up_space = up(x_space) 
down_space = down(x_space) 

q = in_range(y_array, down(x_array), up(x_array))
```

```python
plt.scatter(x_array, y_array)
plt.plot(x_array[q], y_array[q], c = 'green')
plt.plot(x_space, up_space, c= 'red')
plt.plot(x_space, down_space, c = 'blue')
plt.show()
```
